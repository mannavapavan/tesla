```python
# To convert N-D array data into a data frame - collection of 2D
import pandas as pd

# numerical array mathematical operations
import numpy as np

# simple graphs - bar, pie, line plot , histogram
import matplotlib.pyplot as plt

#for distributional graph - heatmap, violinplot, count plot, density plot
import seaborn as sns
```

```python
boston_df = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/boston_train.csv')
```

```python
boston_df.head()
```

```python
boston_df.describe()
```

```python
boston_df.isnull().sum()
```

## Line Plot

```python
plt.figure(figsize=(20,9))
plt.plot(boston_df['AGE'])
plt.title('Line Plot', fontsize = 20)
plt.xlabel('Row Indexes', fontsize = 20)
plt.ylabel('AGE',fontsize = 20)
plt.legend()
```

```python
boston_df['AGE'].sort_values(ascending = True).unique()
plt.figure(figsize=(20,9))
plt.plot(boston_df['AGE'].sort_values(ascending = True).unique())
plt.title('Line Plot', fontsize = 20)
plt.xlabel('Row Indexes', fontsize = 20)
plt.ylabel('AGE',fontsize = 20)
plt.legend()
```

```python
boston_df['AGE'].sort_values(ascending = True).unique()
```

### Histogram

```python
sns.set_style('darkgrid')
plt.hist(boston_df['MEDV'], label = 'MEDV' , bins=30)
#bins is frequency
```

```python
sns.set_style('darkgrid')
plt.hist(boston_df['MEDV'], label = 'MEDV')
plt.title('Histogram', fontsize = 20)
plt.xlabel('MEDV', fontsize = 20)
plt.ylabel('Frequencies of age at each separations',fontsize = 20)
plt.legend()
```

```python
sns.set_style('darkgrid')
plt.hist(boston_df['MEDV'], label = 'MEDV', bins=30)
plt.title('Histogram', fontsize = 20)
plt.xlabel('MEDV', fontsize = 20)
plt.ylabel('Frequencies of age at each separations',fontsize = 12)
plt.legend()
```

```python
iris_df = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/iris.csv')
```

```python
iris_df.head()
```

```python
iris_df.describe()
```

```python
plt.figure(figsize=(12,8))
sns.set_style('darkgrid')
plt.scatter(x = iris_df['sepal_length'], y = iris_df['petal_length'])
plt.scatter(x = iris_df['sepal_length'], y = iris_df['petal_width'])
plt.legend()
```

### Scatter Plot

```python
plt.figure(figsize=(12,8))
sns.set_style('darkgrid')
plt.scatter(x = iris_df['sepal_length'], y = iris_df['petal_length'], label = 'Petal length')
plt.scatter(x = iris_df['sepal_length'], y = iris_df['petal_width'], label = 'Petal width')
plt.legend()
```

### Scatter Plot Seaborn

```python
# scatter plot using seaborn
plt.figure(figsize=(15,8))
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data = iris_df, hue='species')
```

## Density Graph

```python
# scatter plot using seaborn
plt.figure(figsize=(15,8))
sns.distplot(iris_df['sepal_length'], bins = 20, color ='red')
plt.title("Scatter Plot", fontsize = 20)
plt.show()
```

## Box & Whisker Plot (IQR)

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=iris_df)
plt.title("BOX Plot", fontsize = 20)
plt.show()
```
