### Handling Missing Values

```python
titanic_df.info()
```

```python
titanic_df.isnull().sum().sort_values(ascending = False)
```

```python
med_age = titanic_df.loc[:,'Age'].median()
```

```python
med_age
```

```python
# replace missing values (np.nan) with median age value
titanic_df['Age'].replace (np.nan, med_age, inplace=True)
```
